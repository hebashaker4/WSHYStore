 (3) Data Cleaning

```python
null_perc = wshy_store.isnull().sum()/len(wshy_store) *100
print ( 'null percentage per column ' )
print ( null_perc ) 
```

```python
wshy_store['Age'] = wshy_store['Age'].fillna(wshy_store['Age'].median())
wshy_store['Gender'] = wshy_store['Gender'].fillna('Unknown')
wshy_store['Income'] = wshy_store['Income'].fillna(wshy_store['Income'].mode()[0])
```

```python
wshy_store.dropna(subset=['Transaction_ID', 'Customer_ID'], inplace=True)
wshy_store.dropna(subset=['Date', 'Time'] ,  inplace=True)
```

```python
wshy_store['Name'].fillna('Unknown', inplace=True)
wshy_store['Email'].fillna('Unknown', inplace=True)
wshy_store['Phone'].fillna(0, inplace=True)
wshy_store['Address'].fillna('Unknown', inplace=True)
wshy_store['City'].fillna('Unknown', inplace=True)
wshy_store['State'].fillna('Unknown', inplace=True)
wshy_store['Zipcode'].fillna(0, inplace=True)
wshy_store['Country'].fillna('Unknown', inplace=True)
```

```python
wshy_store.columns
```

```python
wshy_store[[ 'Net_Value','DiscountValue','Amount','Total_Amount']]
```

```python
wshy_store['DiscountValue'].unique()
```

```python
discount_mapping = {
    0:1,
    50: 0.5,
    10: 0.9,
    20: 0.8,
    30: 0.7
}


wshy_store['DiscountValue'] = wshy_store['DiscountValue'].replace(discount_mapping)

```

```python
wshy_store['DiscountValue']
```

```python
len(wshy_store)
```

```python
wshy_store['Customer_Segment']=wshy_store['Customer_Segment'].fillna(wshy_store['Customer_Segment'].mode()[0])
```

```python
wshy_store.dropna(subset=['Date', 'Time'], inplace=True)
```

```python
wshy_store['Total_Purchases']=wshy_store['Total_Purchases'].fillna(wshy_store['Total_Purchases'].median())
wshy_store['Amount']=wshy_store['Amount'].fillna(wshy_store['Amount'].mean())
wshy_store['Total_Amount']=wshy_store['Total_Amount'].fillna(wshy_store['Total_Amount'].mean())
```

```python
wshy_store['Product_Category']=wshy_store['Product_Category'].fillna(wshy_store['Product_Category'].mode()[0])
wshy_store['Product_Brand']=wshy_store['Product_Brand'].fillna(wshy_store['Product_Brand'].mode()[0])
wshy_store['Product_Type']=wshy_store['Product_Type'].fillna(wshy_store['Product_Type'].mode()[0])
```

```python
wshy_store['Feedback']=wshy_store['Feedback'].fillna('No Feedback')
wshy_store['Ratings']=wshy_store['Ratings'].fillna('No Rating')
wshy_store['Shipping_Method']=wshy_store['Shipping_Method'].fillna(wshy_store['Shipping_Method'].mode()[0])
wshy_store['Payment_Method']=wshy_store['Payment_Method'].fillna(wshy_store['Payment_Method'].mode()[0])
wshy_store['Order_Status']=wshy_store['Order_Status'].fillna('Unknown')
wshy_store['products']=wshy_store['products'].fillna('Unknown')
wshy_store['BranchName']=wshy_store['BranchName'].fillna('Unknown')
wshy_store['StaffName']=wshy_store['StaffName'].fillna('Unknown')
wshy_store['OfferName']=wshy_store['OfferName'].fillna('No Offer')
wshy_store['DiscountValue']=wshy_store['DiscountValue'].fillna(0)
```

```python
wshy_store['Net_Value']= wshy_store['Net_Value'].fillna(wshy_store['Total_Amount'] * wshy_store['DiscountValue'])
```

```python
null_perct = wshy_store.isnull().sum()/len(wshy_store) *100
print ( 'null percentage per column ' )
print ( null_perct ) 
```

```python
wshy_store.to_csv('cleaned_wshy.csv', index=False)
```

```python
# Replace specific value with NaN
WSHY_STORE_CLEANED['Ratings'] = WSHY_STORE_CLEANED['Ratings'].replace(0, pd.NA)

# Drop rows with NaN in 'Customer_Segment'
WSHY_STORE_CLEANED.dropna(subset=['Ratings'])
```

```python
# Plotting function
def plot_value_counts(value_counts, title, xlabel):
    value_counts.plot(kind='bar', color='skyblue')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Visualize Customer Segment counts
plot_value_counts(customer_segment_counts, 'Customer Segment Distribution', 'Customer Segment')

# Visualize Product Brand counts
plot_value_counts(product_brand_counts, 'Product Brand Distribution', 'Product Brand')

# Visualize Branch Name counts
plot_value_counts(branch_name_counts, 'Branch Name Distribution', 'Branch Name')

# Visualize Offer Name counts

plot_value_counts(offer_name_counts, 'Offer Name Distribution', 'Offer Name')

plot_value_counts(ratings_counts, 'ratings_counts Distribution', 'ratings')
```

4.2 Bivariat Analysis 

```python
country_amount =pd.DataFrame(WSHY_STORE_CLEANED.groupby('Country')['Net_Value'].sum().reset_index())
country_amount
```

```python
plt.figure(figsize=[8,5])
plt.bar(x = country_amount['Country'], height =country_amount['Net_Value'] )
plt.title('Country Total Amount')
plt.xlabel('Country')
plt.ylabel('Total Amount')
plt.show()
```

```python
age_amount = pd.DataFrame(WSHY_STORE_CLEANED.groupby('Age')['Net_Value'].sum().reset_index())
age_amount
```

```python
age_bins = [17, 23, 29, 35, 41, 47, 53, 59, 65, 100]  # Adjust the bin ranges as per your need
age_labels = ['18-23', '24-29', '30-35', '36-41', '42-47', '48-53', '54-59', '60-65', '65+']
```

```python
WSHY_STORE_CLEANED['Age_Group'] = pd.cut(WSHY_STORE_CLEANED['Age'], bins=age_bins, labels=age_labels)
WSHY_STORE_CLEANED['Age_Group']
```

```python
age_amount = WSHY_STORE_CLEANED.groupby('Age_Group', observed=False)['Net_Value'].sum().reset_index().sort_values(by='Net_Value', ascending=False)
plt.figure(figsize=[8, 5])
plt.bar(x=age_amount['Age_Group'], height=age_amount['Net_Value'])
plt.title('Total Net Value by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Net Value')
plt.show()
```
